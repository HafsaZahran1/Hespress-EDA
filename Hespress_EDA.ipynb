```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
import arabicstopwords.arabicstopwords as arabic_stop_words

# Convert arabic_stop_words module to a list
stop_words = list(arabic_stop_words.stopwords_list())

# Replace 'folder_path' with the actual path to the folder containing the CSV files
folder_path = '/content/stories'

# Get a list of all CSV files in the folder that start with 'stories'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv') and file.startswith('stories')]

# Individual File Analysis
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    try:
        data = pd.read_csv(file_path, encoding='utf-8')
    except pd.errors.ParserError as e:
        print(f"Error reading {csv_file}: {e}")
        continue


    print(f"\n--- Individual File Analysis for {csv_file} ---")

    # 1. Data Overview
    print(data.info())

    # 2. Class Distribution
    class_counts = data['topic'].value_counts()
    print("\nClass Distribution:")
    print(class_counts)

    # Plotting Class Distribution
    plt.figure(figsize=(8, 5))
    sns.barplot(x=class_counts.index, y=class_counts.values)
    plt.xlabel('Class')
    plt.ylabel('Number of Examples')
    plt.title('Class Distribution')
    plt.xticks(rotation=45)
    plt.show()

    # 3. Top Frequent n-grams
    n = 2
    vectorizer = CountVectorizer(ngram_range=(n, n), stop_words=None)  # Do not use built-in stop words
    ngram_counts = vectorizer.fit_transform(data['story'])

    # Remove Arabic stop words
    data['story'] = data['story'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop_words))

    ngram_counts = vectorizer.fit_transform(data['story'])
    ngram_freq = ngram_counts.sum(axis=0)
    ngram_freq = [(word, ngram_freq[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    ngram_freq = sorted(ngram_freq, key=lambda x: x[1], reverse=True)

    # Display top 10 frequent n-grams
    print("\nTop 10 Frequent", n, "-grams:")
    for ngram, count in ngram_freq[:10]:
        print(f"{ngram}: {count}")
    # 4. Lengths of Examples
    # a) Length of Examples in Words
    data['num_words'] = data['story'].apply(lambda x: len(x.split()))
    print("\nLength of Examples in Words:")
    print(data['num_words'].describe())

    # Plotting Length of Examples in Words
    plt.figure(figsize=(8, 5))
    sns.histplot(data['num_words'], kde=True)
    plt.xlabel('Number of Words')
    plt.ylabel('Frequency')
    plt.title('Distribution of Length in Words')
    plt.show()

    # b) Length of Examples in Letters
    data['num_letters'] = data['story'].apply(lambda x: len(x))
    print("\nLength of Examples in Letters:")
    print(data['num_letters'].describe())

    # Plotting Length of Examples in Letters
    plt.figure(figsize=(8, 5))
    sns.histplot(data['num_letters'], kde=True)
    plt.xlabel('Number of Letters')
    plt.ylabel('Frequency')
    plt.title('Distribution of Length in Letters')
    plt.show()

```

```python
# Overview Analysis
print("\n--- Overview Analysis of All 'stories' Files ---")
# Concatenate all 'stories' files into a single DataFrame
all_data = pd.DataFrame()
for csv_file in csv_files:
  file_path = os.path.join(folder_path, csv_file)
  data = pd.read_csv(file_path, encoding='utf-8')
  all_data = pd.concat([all_data, data])

# Reset the index of the combined DataFrame
all_data.reset_index(drop=True, inplace=True)

# 1. Data Overview
print(all_data.info())

# 2. Class Distribution
class_counts = all_data['topic'].value_counts()
print("\nClass Distribution:")
print(class_counts)

# Plotting Class Distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.xlabel('Class')
plt.ylabel('Number of Examples')
plt.title('Class Distribution')
plt.xticks(rotation=45)
plt.show()

# 3. Top Frequent n-grams
n = 2
vectorizer = CountVectorizer(ngram_range=(n, n), stop_words=stop_words)  # Use the Arabic stop words
ngram_counts = vectorizer.fit_transform(all_data['story'])
ngram_freq = ngram_counts.sum(axis=0)
ngram_freq = [(word, ngram_freq[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
ngram_freq = sorted(ngram_freq, key=lambda x: x[1], reverse=True)

# Display top 10 frequent n-grams
print("\nTop 10 Frequent", n, "-grams:")
for ngram, count in ngram_freq[:10]:
  print(f"{ngram}: {count}")

# 4. Lengths of Examples
# a) Length of Examples in Words
all_data['num_words'] = all_data['story'].apply(lambda x: len(x.split()))
print("\nLength of Examples in Words:")
print(all_data['num_words'].describe())

# Plotting Length of Examples in Words
plt.figure(figsize=(8, 5))
sns.histplot(all_data['num_words'], kde=True)
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.title('Distribution of Length in Words')
plt.show()

# b) Length of Examples in Letters
all_data['num_letters'] = all_data['story'].apply(lambda x: len(x))
print("\nLength of Examples in Letters:")
print(all_data['num_letters'].describe())

# Plotting Length of Examples in Letters
plt.figure(figsize=(8, 5))
sns.histplot(all_data['num_letters'], kde=True)
plt.xlabel('Number of Letters')
plt.ylabel('Frequency')
plt.title('Distribution of Length in Letters')
plt.show()
```
